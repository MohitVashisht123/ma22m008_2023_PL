```python
import numpy as np
import matplotlib.pyplot as plt

# Define paramaters
# Number of grid points
n = 2000  
 # Grid spacing
h = 10.0 / (n - 1) 
x = np.linspace(0, 10, 1999) 

y = np.linspace(0, 10, 2000) # Equally spaced values from 0 to 10
# Adjusting the length of matrix to match dimensions 
print(len( np.diag(-np.ones(1998), k=-1) ))
# Create the differentiation matrix D
D = (1 / h) * (np.diag(-np.ones(n-2), k=-1) + np.diag(np.ones(n-1), k=0))

# Define the functions
def f1(x):
    return np.sin(x)

def f2(x):
    return x**2 + 2 * np.sin(10 * x)

# Apply D to approximate the derivatives
df1 = (1 / h) * np.dot(D, f1(x))
df2 = (1 / h) * np.dot(D, f2(x))

# Plot the results
plt.figure(figsize=(20, 6))

plt.subplot(2, 1, 1)
plt.plot(y[:-1], df1)
plt.title(" df1/dx")
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(y[:-1], df2)
plt.title(" df2/dx")
plt.grid()


plt.show()
```

```python
import numpy as np
def twice(w):
    w*=2
    return w
v=[1,2,3]
print("New value",twice(v))
print("Original value",v)
# This happens since python pass the object by reference as an argument. It modifies the list in place.This could be prevented by making deep copy
```

```python
import numpy as np
import copy
def twice(w):
    lst=copy.deepcopy(w)
    lst*=2
    return lst
v=[1,2,3]
print("New value",twice(v))
print("Original value",v)
```

```python
v="Mohit"
print(twice(v))
# Same is the case with the string values
```

```python
import matplotlib.pyplot as plt
import math
def f(x):
    return x*(np.sin(x**2))
# Defining the function
x=np.linspace(-5,5,1000001)
# Generating the x values using linspace
y=f(x)


# Plot the x*sin(x^2) function
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of x*sin(x^2)')

plt.show()


```

```python
import matplotlib.pyplot as plt
import math
def f(x):
    return np.cos(x)
x=np.linspace(-5,5,1000001)
y=f(x)
plt.plot(x,y)
# Plot the cos(x) function
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of cos(x)')

plt.show()
```

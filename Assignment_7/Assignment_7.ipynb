```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Define the ODE system
def ode_system(t, y, a, b, c, d):
    r,x=y[0],y[1]
    r_diff= a* r - b * r * x
    t_diff = c * r * x - d* x
    return [r_diff, t_diff]
# Setting the parameters
a = 1.1
b = 0.4
c= 0.1
d = 0.1
# Setting the initial conditions
initial_conditions = [5, 2]

# Defining the time span
t_span = (0, 100)





# Solve the ODE system
sol = solve_ivp(
    ode_system,
    t_span,
    initial_conditions,
    args=(a, b, c, d),
    t_eval=np.linspace(0, 70, 1000)
)



# Solution extract

t_values, (r_values, x_values) = sol.t, sol.y






# Plot the solutions
plt.figure(figsize=(12, 6))
plt.plot(t_values, r_values, label='Rabbits (r)')
plt.plot(t_values, x_values, label='Foxes (x)')
plt.xlabel('Time')

plt.ylabel('Pop')

plt.title('Rabbits and Foxes')

plt.show()
```

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Defining the ODE system
def ode_system(y, t, a, b, c, d):
    r, x = y[0],y[1]
    drdt = a * r - b * r * x
    dxdt = c * r * x - d * x
    return [drdt, dxdt]





# Setting the parameters
a = 1.1
b = 0.4
c = 0.1
d = 0.1
# Defining the time span
t = np.linspace(0, 100, 1000)
# Set the initial conditions
ini = np.random.uniform(1, 9, size=(10, 2))





# Creating the phase plot
plt.figure(figsize=(12, 6))

i = 0
while i < len(ini):
    r0, x0 = ini[i]
    y0 = [r0, x0]
    solution = odeint(ode_system, y0, t, args=(a, b, c, d))
    plt.plot(solution[:, 1], solution[:, 0], label=f'r0={r0:.2f}, x0={x0:.2f}')
    i += 1

plt.xlabel('Foxes (x)')

plt.ylabel('Rabbits (r)')

plt.title('Phase Portrait of Rabbit-Fox ')

plt.legend()

plt.show()
```

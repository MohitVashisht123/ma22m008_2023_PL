```python
import numpy as np

import matplotlib.pyplot as plt

k=3
def k_nearest_neighbors(points, k):
#     Function definition
    n = points.shape[0]

    dist=np.zeros((n,n))
    ind=np.zeros((n,k),dtype=int)
    i=0
    for i in range(n):
        for j in range(n):
            if i!=j:
                diff=p[i]-p[j]
                dist[i,j]=np.linalg.norm(diff)
#                 Calculation of euclidean dist
            else:
                dist[i,j]=np.inf
#             dist of point to itself
    for i in range (n):
        ind_k=np.argpartition(dist[i],k)[:k]
#         calculation of indices of nearest neighbours
    #     print(ind_k)
        ind[i]=ind_k
        dist[i,ind_k]=dist[i,ind_k]
    return ind
    # print(ind)
p = np.array([[0,0], [0.2, 0.22], [0.1, -0.1],
              [1,1], [1.1, 0.9], [0.8, 0.9], [1.1, 0.63],
              [0.58, -0.1], [0.63, 0.1], [0.67, -0.3], [0.8,-0.23],
              [0.8, 0.6]])
k = 3


nearest_neighbors_indices = k_nearest_neighbors(p, k)
#  k-nearest neighbors for each point


plt.scatter(p[:, 0], p[:, 1])
# Ploting the points
# Plot arrows from each point to its k-nearest neighbors
i = 0
j = 0


while i<len(p):
    j=0
    while j < len(nearest_neighbors_indices[i]):
        if i != nearest_neighbors_indices[i][j]:
            plt.arrow(p[i, 0],p[i, 1],p[nearest_neighbors_indices[i][j], 0] -p[i, 0],p[nearest_neighbors_indices[i][j], 1] -p[i, 1], head_width=0.05, head_length=0.1, fc='r', ec='r')
        j += 1
    
    i += 1
# Plotting of arrows
# Set axis limits
plt.xlim(-0.5, 1.5)
plt.ylim(-0.5, 1.5)

# Show the plot
plt.show()


```

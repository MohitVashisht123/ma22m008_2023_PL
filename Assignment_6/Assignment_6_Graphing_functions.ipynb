```python
import numpy as np
import matplotlib.pyplot as plt

# Generating the points
x = np.linspace(0, 50, 1000)

# Calculating the corresponding y values
y = np.sin(x)

# Create the plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='sin(x)', color='red')
plt.title('Graph of sin(x) ')
plt.xlabel('x')
plt.ylabel('sin(x)')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Generating the points
x=np.linspace(-25,25,500)
y=np.linspace(-25,25,500)
# Making the grid
X,Y=np.meshgrid(x,y)
Z=np.cos(X**2+Y**2)
figure=plt.figure(figsize=(8,6))
axes=figure.gca(projection='3d')
surf=axes.plot_surface(X,Y,Z,cmap='viridis',linewidth=0)
# Axes label
axes.set_xlabel('x axis')
axes.set_ylabel('y axis')
axes.set_zlabel('cos(x^2 +y^2')
# Ploting
plt.show()
```
